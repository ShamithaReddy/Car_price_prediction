```python
#External TreeRegressor gives you feauture importance
from sklearn.ensemble import ExtraTreesRegressor
```

```python
from sklearn.model_selection import train_test_split
```

```python
X_train,X_test,y_train,y_test=train_test_split(X,y,test_size=0.33)
```

```python
X_train.shape
```

```python
X_test.shape
```

```python
from sklearn.ensemble import RandomForestRegressor
```

```python
regressor=RandomForestRegressor() 
#n_estimators--->decision trees
```

```python
#Hyperparameters
n_estimators=[i for i in range(100,1201,100)]
```

```python
print(n_estimators)
# different no. of trees in random forest
```

```python
# Number of features to consider at every split
#If “auto”, then max_features=n_features.
#If “sqrt”, then max_features=sqrt(n_features).
max_features = ['auto', 'sqrt']
```

```python
# Maximum number of levels in tree
max_depth = [int(x) for x in range(5,31,5)]
print(max_depth)
```

```python
# Minimum number of samples required to split a node
min_samples_split = [2, 5, 10, 15, 100]
```

```python
# Minimum number of samples required at each leaf node
min_samples_leaf = [1, 2, 5, 10]
```

```python
from sklearn.model_selection import RandomizedSearchCV
```

```python
#randomizedsearchcv selects out the best paramenters out of all parameters in random grid
random_grid={"n_estimators":n_estimators,
             "max_depth":max_depth,
             "max_features":max_features,
             "min_samples_split":min_samples_split,
             "min_samples_leaf":min_samples_leaf}
print(random_grid)
# We use the random grid to search for best hyperparameters
```

```python
# First create the base model to tune
rf=RandomForestRegressor()
```

```python
# Random search of parameters, using 3 fold cross validation, 
# search across 100 different combinations
rf_random = RandomizedSearchCV(estimator = rf, param_distributions = random_grid,scoring='neg_mean_squared_error', n_iter = 10, cv = 5, verbose=2, random_state=42, n_jobs = 1)
```

```python
rf_random.fit(X_train,y_train)
```

```python
rf_random.best_params_
```

```python
prediction=rf_random.predict(X_test)
```

```python
plt.scatter(y_test,prediction)
```

```python
sns.distplot(y_test-prediction)
# it is giving a normal distribution, so it means the results are good
```

```python
from sklearn import metrics
```

```python
print("MSE",metrics.mean_squared_error(y_test,prediction))
print("MAE",metrics.mean_absolute_error(y_test,prediction))
```
